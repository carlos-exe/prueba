#**Gráficas y ecuaciones**

Librería: matplotlib.pyplot


```python
#Code
import matplotlib.pyplot as plt

x = list(range(100))
y = [i for i in x]

plt.plot(x,y)
plt.title("Gráfica de Y = X")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Y = X+3
x = list(range(100))
y = [i+3 for i in x]

plt.plot(x,y)
plt.title("Gráfica de Y = X+3")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

$$Y = x^2$$

```python
# Y = X^2
import numpy as np

x = np.arange(-100,100,0.1)
y = [i**2 + 2*i + 1 for i in x]

plt.plot(x,y)
plt.title("Gráfica de Y = X^2")
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim(-2.5,0.5)
plt.ylim(-2,2)
plt.show()
```
